```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def get_lambda (v):
    rou = 0.05
    v_max = 35
    Lambda = rou*v*(1-v/v_max)
    return Lambda
```

```python
get_lambda(10),get_lambda(20),get_lambda(30)
```

```python
x = np.linspace(0,30,100)
y = get_lambda(x)
plt.plot(x,y)
```

```python
def get_E(v):
    Lambda = get_lambda(v)
    a = 0.2
    u = 1/0.2
    Rc = 420
    v = v
    T_stay = Rc/v
    D = 2
    #T_round = 40
    T_round = 100
    H = 15

    tau = np.min([T_round,T_stay])
    p1 = 1/(tau)*(tau-a*H-D-H/u*(1-np.exp(-u/H*(tau-a*H-D))))
    p2 = 1-np.exp(-u/H*(tau-a*H-D))
    assert T_stay > a*H+D
    assert T_round > a*H+D

    E = 2*Lambda*tau*p1 + Lambda*np.abs(T_round-T_stay)*p2
    return E
```

```python
x = np.linspace(10,20,3)
y = [get_E(v) for v in x]
plt.plot(x,y)
plt.xlabel('speed')
plt.ylabel('avg car num')
```
